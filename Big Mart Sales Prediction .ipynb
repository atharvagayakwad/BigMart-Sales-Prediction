# Filling Missing Values

### Item Weight is numeric value so we fill it with Mean value

```python
item_wt_mean = df_train['Item_Weight'].mean()
df_train['Item_Weight'].fillna(item_wt_mean, inplace=True)
df_test['Item_Weight'].fillna(item_wt_mean, inplace=True)
```

```python
df_train.isnull().sum()
```

### Outlet Size is categorical value so we fill it with Mode value

```python
outlet_size_mode = df_train['Outlet_Size'].mode()[0]
df_train['Outlet_Size'].fillna(outlet_size_mode, inplace=True)
df_test['Outlet_Size'].fillna(outlet_size_mode, inplace=True)
```

# Splitting Data for Training and Testing

```python
df_train
```

```python
X = df_train.drop('item_outlet_sales', axis=1)
y = df_train['item_outlet_sales']
```

# Train | Test split 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)  
```

# Normalization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

# Model Building

# 1) Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(scaled_X_train, y_train)
```

```python
linear_preds = model.predict(scaled_X_test)
linear_preds
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error  
```

```python
r2_score(y_test, linear_preds)
```

```python
mean_absolute_error(y_test, linear_preds)
```

```python
np.sqrt(mean_squared_error(y_test, linear_preds))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=1000, max_depth=6)
rfr.fit(X_train, y_train)
```

```python
rf_preds = rfr.predict(X_test)
rf_preds
```

```python
r2_score(y_test, rf_preds)
```

```python
mean_absolute_error(y_test, rf_preds)
```

```python
np.sqrt(mean_squared_error(y_test, rf_preds))
```

# Hyper-parameter Tuning 

```python
from sklearn.model_selection import GridSearchCV

n_estimators = [10,100,1000]
max_depth = range(1,11)


param_grid = {'n_estimators':n_estimators,
              'max_depth':max_depth,
             }

rfr = RandomForestRegressor()
grid_model = GridSearchCV(rfr, param_grid=param_grid)
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_params_
```

```python
grid_model.best_score_
```
